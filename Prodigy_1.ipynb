```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv('age_gender.csv')
x
```

```python
print("No.of Rows: {}".format(len(x)))
print("No.of Columns: {}".format(len(x.columns)))
```

```python
# Bar graph for Age Distribution
age_dist = x['age'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(age_dist.index, age_dist.values, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Bar graph for Gender Distribution
gender_dist = x['gender'].value_counts().rename(index={0: 'Male', 1: 'Female'})
plt.figure(figsize=(8, 6))
plt.bar(gender_dist.index, gender_dist.values, color='pink')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Bar graph for Ethnicity Distribution
ethnicity_dist = x['ethnicity'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(ethnicity_dist.index, ethnicity_dist.values, color='lightgreen')
plt.title('Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for Age
plt.figure(figsize=(10, 6))
plt.hist(x['age'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
